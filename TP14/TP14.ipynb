```python
reset()
print("""\
# ****************************************************************************
# PAS DE BEBE, PAS DE GEANT
# ****************************************************************************
""")


# Donnees de l'enonce de l'exercice

p1 = 1823
Fp1 = FiniteField(p1)
b1 = Fp1(3)
x1 = Fp1(693)

p2 = 239
Fp2 = FiniteField(p2)
b2 = Fp2(2)
x2 = Fp2(15)


# Code pour l'EXERCICE

def Shanks(g,b):
    G = g.parent()
    n = G.order()
    s = floor(sqrt(n)) + 1
    T = {}
    
    for j in range(s):
        betaj = g*b^(-j)
        T[betaj] = j
    
    i = 0
    gamma = G.one()
    bp = b^s
    found = False
    
    while not found:
        i+=1
        gamma = gamma*bp
        if T.has_key(gamma):
            found = True
    j = T[gamma]
    return i*s + j


# # Affichage des resultats
k1 = Shanks(x1,b1)
k2 = Shanks(x2,b2)
print "Question 2 :", k1, " | Check : b^k = ", b1^k1 , " x = ", x1
print "Question 3 :", k2, " | Check : b^k = ", b2^k2 , " x = ", x2

```
