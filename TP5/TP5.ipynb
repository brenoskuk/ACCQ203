```python
F3 = FiniteField(3)
Pol3.<x> = PolynomialRing(F3)
f = x^3 - x^2 - 1


def myB(f):
    Pol=f.parent()
    x=Pol.gen()
    p=Pol.base_ring().characteristic()
    q=Pol.base_ring().cardinality()
    retour = [f]
    # ligne 1
    n = f.degree()
    Q = matrix(Pol,n,n)
    for j in range(n):
        g = x**(j*n)
        gg = (g%f)
        # il faut completer gg jusqu'a n
        for v in range(len(gg.list())): 
            Q[v,j] = gg.list()[v]
    
    # ligne 3 - calcul d'une base du noyau de Q - In 
    Ker = (Q-matrix.identity(Pol,n)).right_kernel()  
    k = len(Ker.gens())
    base = matrix(Pol,k,n)
    for i in range(k):
        e = Ker.gen(i)
        bi = [e[j]*x**j for j in range(len(e))]
        base[i] = bi
    
    # ligne 6
    j = 0    
    while (len(retour)<k):
        j = j + 1
        C = [f for f in retour if f.degree() > 1]
        for f in C:
            B = []
            for alpha in GF(p):
                print 'f = ',f,'bj = ', base[j], 'alpha = ', alpha, '\n'
                a = xgcd(f,(base[j] - alpha))
                if a.degree() >= 1:
                    B = B.append(a)
            retour.remove(f)
            retour = retour + B
#    assert(Set(retour) == Set(g for g,_ in list(f.factor())))
    return retour
```
