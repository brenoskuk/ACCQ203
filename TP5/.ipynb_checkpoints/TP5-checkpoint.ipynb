```python
print("""\
# *************************************************************************** #
# *************************************************************************** #
# TP4 : FACTORISATION COMPLETE DE POLYNOMES UNIVARIEES                        #
# *************************************************************************** #
# *************************************************************************** #
""")

# CONSIGNES
#
# Les seules lignes a modifier sont annoncee par "Code pour l'exercice"
# indique en commmentaire et son signalees
# Ne changez pas le nom des variables
#
# CONSEILS
#
# Ce modele vous sert a restituer votre travail. Il est deconseille d'ecrire
# une longue suite d'instruction et de debugger ensuite. Il vaut mieux tester
# le code que vous produisez ligne apres ligne, afficher les resultats et
# controler que les objets que vous definissez sont bien ceux que vous attendez.
#
# Vous devez verifier votre code en le testant, y compris par des exemples que
# vous aurez fabrique vous-meme.
#


reset()
print("""\
# ****************************************************************************
# FACTORISATION DES PUISSANCES
# ****************************************************************************
""")


# Donnees de l'enonce de l'exercice

F3 = FiniteField(3)
Pol3.<x> = PolynomialRing(F3)
f = x^3 - x^2 - 1

# Code pour l'EXERCICE

x3 = Pol3(0)
x6 = Pol3(0)
Q = matrix(F3,3)

b1 = vector(F3,[1,0,0])
b2 = vector(F3,[0,1,1])

L = []

def myB(f):
    Pol=f.parent()
    x=Pol.gen()
    p=Pol.base_ring().characteristic()
    q=Pol.base_ring().cardinality()
    F = [f]
    # ligne 1
    n = f.degree()
    Q = matrix(Pol,n,n)
    for j in range(n):
        g = x**(j*n)
        gg = (g%f)
        # il faut completer gg jusqu'a n
        for v in range(len(gg.list())): 
            Q[v,j] = gg.list()[v]
    # ligne 3 - calcul d'une base du noyau de Q - In 
    Ker = (Q-matrix.identity(Pol,n)).right_kernel()  
    k = len(Ker.gens())
    base = []
    for i in range(k):
        e = Ker.gen(i)
        bi = 0
        for j in range(len(e)):
            bi = bi + e[j]*x**j
        base.append(bi)    
    # ligne 6
    j = 0    
    #return base
    while (len(F)<k):
        j = j + 1
        C = [f for f in F if f.degree() > 1]
        for f in C:
            B = []
            for alpha in Pol.base_ring():
                # print 'f = ',f,'bj = ', base[j], 'alpha = ', alpha, '\n'
                a,_,_ = xgcd(f,(base[j] - alpha)) 
                if a.degree() >= 1:
                    B.append(a)
            F.remove(f)
            F = F + B
    assert(Set(F) == Set(g for g,_ in list(f.factor())))
    retour = F
    return retour

# Pas d'implementation dans le dernier TP
def myFsFC(L):
    exposants = []
    return L,exposants

def myFactor(f):
    Pol=f.parent()
    x=Pol.gen()
    p=Pol.base_ring().characteristic()
    q=Pol.base_ring().cardinality()
    L = myFsFC(f)
    # Decoupage par exposants
    U=[]
    for i in range(len(L)):
        # Decoupage par Berlekamp de chaque composant
        U.append((myB(L[i]),exposants))
    return U
        
    retour = [(f,1)]
#    assert(Set(retour) == Set(list(f.factor())))
    return retour


# # Affichage des resultats
## Tests done with assert

F4 = FiniteField(4)
Pol4.<x> = PolynomialRing(F4)
f1 = x^4 + x^3 + x^2 + 1
F5 = FiniteField(5)
Pol5.<x> = PolynomialRing(F5)
f2 = x^5 + 2*x^4 + x + 1
F7 = FiniteField(7)
Pol7.<x> = PolynomialRing(F7)
f3 = 6*x^7 + 5*x^6 + 2*x^5 + 5*x^4 + 3*x^3 + 6*x^2 + 6*x + 5 
F17 = FiniteField(17)
Pol17.<x> = PolynomialRing(F17)
f4 = 11*x^17 + 3*x^16 + 6*x^15 + 10*x^14 + 5*x^13 + 6*x^12 + 16*x^11 + 15*x^10 + 14*x^9 + 12*x^8 + 14*x^7 + 6*x^6 + 7*x^4 + x^3 + 3*x^2 + 8*x 


fs = []
fs.append(f)
fs.append(f1)
fs.append(f2)
fs.append(f3)
fs.append(f4)
results = []
results.append(myB(f))
results.append(myB(f1))
results.append(myB(f2))
results.append(myB(f3))
results.append(myB(f4))

print '** Tests **'
for i in range(len(results)):
    print '\n',fs[i],' =? ',results[i],'\n'
    if (Set(results[i]) == Set(g for g,_ in list(fs[i].factor()))):
        print '...Yes\n!'
    else: 
        print '...No'
        
reset()
print("""\
# ****************************************************************************
# RELEVEMENT DE HENSEL
# ****************************************************************************
""")


# Donnees de l'enonce de l'exercice

PolZZ.<x> = PolynomialRing(ZZ)
m = 5
f = x^4-1
g = x^3+2*x^2-x-2
h = x-2
d,ss,tt = xgcd(g,h)
s=PolZZ(ss/mod(d,m)); t=PolZZ(tt/mod(d,m))

# Code pour l'EXERCICE

def SymMod(a,m):
    a=mod(a,m).lift()
    if a<=m/2:
        return a
    else:
        return a-m

def mod_polyn(f,m):
    return f.parent()(map(lambda x:SymMod(x,m), f))

def polynomeCentre(f,m):
    Pol=f.parent()
    l = f.list()
    x = Pol.gen()
    ln = []
    for i in range(len(l)):
        ln.append(Mod(l[i],m).lift_centered())
    return Pol(ln)


def myHensel(f,r,p,k):
    for i in range(0,k):
        x = mod_polyn(f(r)*p^(-2**i),p^(2**i))
        z = mod_polyn(f.derivative()(r)^(-1),p^(2**i))
        y = mod_polyn(-x*z,p^(2**i))
        r = r + y*p^(2**i)
    return r

def myHenselItere(f,g,h,p,l):
    Pol=f.parent()
    x=Pol.gen()
    # f - g^h^ == 0 mod pl
    # Relevement de Hensel trouve r tq
    # ff == 0 mod p**(2^k)
    myHensel(f-gh,)
    t = 0
    while(2^t<l):
        t = t + 1
        
    retour = []
    return retour

reponseQ5=""

# # Affichage des resultats

print "\n$1b/ Relèvement de ",f,"= (",g,")*(",h,")"
#print myHensel(f,g,h,s,t,m)
print reponseQ5

reset()
print("""\
# ****************************************************************************
# FACTORISATION AVEC LLL
# ****************************************************************************
""")


# Donnees de l'enonce de l'exercice

p=13
k=4
m=p^k

PolZZ.<x> = PolynomialRing(ZZ)
f = x^4-x^3-5*x^2+12*x-6

alpha=0
beta=0
gamma=0
delta=0
alphahat=0
betahat=0
gammahat=0
deltahat=0

u = x+7626

# Code pour l'EXERCICE




# # Affichage des resultats

print "\n$1a/ Les racines sont", alpha, beta, gamma, delta,"modulo",p
print "\n$1b/ Les racines sont", alphahat, betahat, gammahat, deltahat,"modulo",m
```
